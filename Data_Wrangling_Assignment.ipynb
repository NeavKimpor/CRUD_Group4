```python
import sklearn #marchine learing
import numpy as np #computation
import pandas as pd #data management
```

```python
#1. Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/Opensourcefordatascience/Data-sets/master/blood_pressure.csv")
dataset.head(10)
```

```python
#2. Dimension
dataset.shape
```

```python
#3. Check the variable(Numberical / Categorical)
#Check numerical data
numerical_variable = dataset.columns[dataset.dtypes != 'object']
dataset[numerical_variable].head(5)
```

```python
#Check categorical data
categorical_variable = dataset.columns[dataset.dtypes == 'object']
dataset[categorical_variable].head(5)
```

```python
#4. Identity the missing value
dataset.isnull()
```

```python
#numerical_value
dataset[numerical_variable].isnull().head(10)
```

```python
#categorical_value
dataset[categorical_variable].isnull().head(10)
```

```python
#5. Find out the percentage of missing value
dataset.isnull().sum()/len(dataset)
```

```python
#6. Replace the missing value
dataset.fillna("No Missing Value")
```

```python
#7. Fill up the missing value with specific variable
dataset['sex'].replace("Male", "Female")
```

```python
#8. Create the row and add it with the original dataset
row_insert = pd.Series([121, 'Male', 62, 170, 183], index = ["patient", "sex", "agegrp", "bp_before", "bp_after"])
dataset.append(row_insert, ignore_index=True)
```

```python
#9. Call the exact row from the dataset
dataset.iloc[0]
dataset.iloc[1:10]
dataset.iloc[:13]
```

```python
#Call the specific value
dataset = dataset.set_index(dataset["sex"])
dataset.loc['Male'].head(5)
```

```python
#10. Rename the columns
dataset.rename(columns = {'sex' : 'gender'}).head(5)
```

```python
#11. Find the minimum and maximum, sum, count
dataset.min()
dataset.max()
dataset.sum()
dataset.count()
```

```python
#12. Delete the row and column
# delete the first column
dataset.drop('bp_before', axis = 1).head(5)
```
